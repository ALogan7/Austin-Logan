### 4)

```python
titanic = Table.read_table("train.csv")
titanic
```

```python
titanic.relabel('Pclass','passenger_class')
```

```python
titanic.append_column('family_size',titanic.column('SibSp')+titanic.column('Parch'))
titanic
```

```python
sort_class = titanic.group("passenger_class")
sort_class
```

```python
firstclasstotal = sort_class.column("count").item(0)
secondclasstotal = sort_class.column("count").item(1)
thirdclasstotal = sort_class.column("count").item(2)
```

```python
sort_class.barh("passenger_class")
```

```python
titanic.sort("Age")
```

```python
titanic.sort("Age",descending=True).show(200)
```

```python
titanic.sort("Age",descending=True).exclude(range(177))
```

```python
surv_rate = titanic.select("Survived","passenger_class")
surv_rate_grouped = surv_rate.group("passenger_class",collect=sum)
surv_rate_grouped
```

```python
firstsurv = surv_rate_grouped.column("Survived sum").item(0)
secondsurv = surv_rate_grouped.column("Survived sum").item(1)
thirdsurv = surv_rate_grouped.column("Survived sum").item(2)
print("The survival rate for the first class is " + str(firstsurv/firstclasstotal))
print("The survival rate for the second class is " + str(secondsurv/secondclasstotal))
print("The survival rate for the third class is " + str(thirdsurv/thirdclasstotal))
```
